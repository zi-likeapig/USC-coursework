### i. Research exact match and hamming score/ loss methods for evaluating multilabel classification and use them in evaluating the classifiers in this problem.

#### Exact Match Ratio

which is also called Subset Accuracy. It calculates the proportion of samples where all labels are correctly predicted. It is a strict measure and only counts a sample as correct if every label matches exactly.

$$
\text{Exact Match Ratio} = \frac{1}{N} \sum_{i=1}^N \mathbf{1}(y_i = \hat{y}_i)
$$

- $N$ is the total number of samples.  
- $y_i$ and $\hat{y}_i$ are the true and predicted label sets for the $i$-th sample.  
- The indicator function $\mathbf{1}(y_i = \hat{y}_i)$ returns 1 if all labels match, and 0 otherwise.


#### Hamming Score

which is alse called Complement of Hamming Loss. It measures how many individual labels are predicted correctly, regardless of whether all labels are correct. It is more tolerant than Exact Match.

$$
\text{Hamming Loss} = \frac{1}{N \cdot L} \sum_{i=1}^N \sum_{j=1}^L \mathbf{1}(y_{ij} \ne \hat{y}_{ij}) \quad \Rightarrow \quad \text{Hamming Score} = 1 - \text{Hamming Loss}
$$

- $L$ is the number of labels per sample (here, 3: Family, Genus, Species).  
- It averages the fraction of incorrect labels across all samples and labels.

```python
def exact_match_ratio(y_true,y_pred):
    return np.mean(np.all(y_true==y_pred,axis=1))

def hamming_score(y_true,y_pred):
    return 1-np.mean(y_true!=y_pred)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
```

```python
X=df.iloc[:,:22].values

y_family=LabelEncoder().fit_transform(df["Family"])
y_genus=LabelEncoder().fit_transform(df["Genus"])
y_species=LabelEncoder().fit_transform(df["Species"])
y_all=np.stack([y_family, y_genus, y_species], axis=1)

X_train,X_test,y_train_all,y_test_all=train_test_split(X,y_all,test_size=0.3,random_state=42,stratify=y_all[:,2])

scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)

print(X_train_scaled.shape)
print(y_test_all.shape)
```

```python
def train_svm_for_label(y_train, X_train, param_grid):
    model=OneVsRestClassifier(SVC(kernel='rbf'))
    clf=GridSearchCV(model,param_grid,cv=10,n_jobs=-1,verbose=0)
    clf.fit(X_train,y_train)
    return clf

param_grid={
  'estimator__C': [1e-3, 1e-2, 1e-1, 1, 10, 100],
  'estimator__gamma': [1e-3, 1e-2, 1e-1, 1]
}

clf_family=train_svm_for_label(y_train_all[:,0],X_train_scaled,param_grid)
clf_genus=train_svm_for_label(y_train_all[:,1],X_train_scaled,param_grid)
clf_species=train_svm_for_label(y_train_all[:,2],X_train_scaled,param_grid)
```

```python
y_pred_family=clf_family.predict(X_test_scaled)
y_pred_genus=clf_genus.predict(X_test_scaled)
y_pred_species=clf_species.predict(X_test_scaled)

y_pred_all=np.stack([y_pred_family, y_pred_genus, y_pred_species], axis=1)

print("Exact Match Ratio:", exact_match_ratio(y_test_all, y_pred_all))
print("Hamming Score    :", hamming_score(y_test_all, y_pred_all))
```
