```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#  ( a)  b) )
#  a)
A = np.array([[2, 1, -1],
              [1, 4, 1],
              [-1, 1, 4]])

# 
eigenvalues, eigenvectors = np.linalg.eigh(A)
print(eigenvalues)
print(eigenvectors)

# 
lengths = 1 / np.sqrt(eigenvalues)  # 
print(lengths)
directions = eigenvectors

# 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 50)
x = lengths[0] * np.outer(np.cos(u), np.sin(v))
y = lengths[1] * np.outer(np.sin(u), np.sin(v))
z = lengths[2] * np.outer(np.ones_like(u), np.cos(v))

# 
ellipsoid = np.dot(np.array([x.ravel(), y.ravel(), z.ravel()]).T, directions.T)
x_r = ellipsoid[:, 0].reshape(x.shape)
y_r = ellipsoid[:, 1].reshape(y.shape)
z_r = ellipsoid[:, 2].reshape(z.shape)

# 
ax.plot_surface(x_r, y_r, z_r, color='cyan', rstride=4, cstride=4, alpha=0.6)
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
plt.show()
```
