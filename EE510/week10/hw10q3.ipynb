```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 设定方程中的系数矩阵 (根据你的方程 a) 和 b) 设置)
# 例如，对于方程 a)：
A = np.array([[2, 1, -1],
              [1, 4, 1],
              [-1, 1, 4]])

# 找到特征值和特征向量
eigenvalues, eigenvectors = np.linalg.eigh(A)
print(eigenvalues)
print(eigenvectors)

# 主轴长度和方向
lengths = 1 / np.sqrt(eigenvalues)  # 长度与特征值的倒数平方根成正比
print(lengths)
directions = eigenvectors

# 绘制椭球
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 使用参数方程绘制椭球
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 50)
x = lengths[0] * np.outer(np.cos(u), np.sin(v))
y = lengths[1] * np.outer(np.sin(u), np.sin(v))
z = lengths[2] * np.outer(np.ones_like(u), np.cos(v))

# 将椭球旋转到主轴方向
ellipsoid = np.dot(np.array([x.ravel(), y.ravel(), z.ravel()]).T, directions.T)
x_r = ellipsoid[:, 0].reshape(x.shape)
y_r = ellipsoid[:, 1].reshape(y.shape)
z_r = ellipsoid[:, 2].reshape(z.shape)

# 绘制椭球
ax.plot_surface(x_r, y_r, z_r, color='cyan', rstride=4, cstride=4, alpha=0.6)
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
plt.show()
```
